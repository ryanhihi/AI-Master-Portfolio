# Python and Data Fundamentals

This notebook reviews core Python, NumPy, Pandas, and Matplotlib concepts used in machine learning.

```python
name = "Ryan"
years_of_experience = 2
is_ai_student = True

skills = [
    "Python",
    "JavaScript",
    "Machine Learning",
    "Full-Stack Development"
]

profile = {
    "name": name,
    "experience_years": years_of_experience,
    "is_ai_student": is_ai_student,
    "skills": skills
}

print(profile)
```

```python
def calculate_average(values: list[float]) -> float:
    if not values:
        raise ValueError("The values list cannot be empty.")

    return sum(values) / len(values)


scores = [82.5, 91.0, 88.5, 95.0]

average_score = calculate_average(scores)

print(f"Average score: {average_score:.2f}")
```

```python
import numpy as np

scores_array = np.array([82.5, 91.0, 88.5, 95.0])

print("Array:", scores_array)
print("Mean:", scores_array.mean())
print("Minimum:", scores_array.min())
print("Maximum:", scores_array.max())
print("Standard deviation:", scores_array.std())
```

```python
adjusted_scores = scores_array + 2

print("Original scores:", scores_array)
print("Adjusted scores:", adjusted_scores)
```

```python
import pandas as pd

student_data = {
    "student": ["Ryan", "Alex", "Jordan", "Taylor"],
    "python_score": [95, 82, 88, 91],
    "ml_score": [90, 79, 92, 87],
    "hours_studied": [12, 7, 15, 10]
}

students_df = pd.DataFrame(student_data)

students_df
```

```python
students_df.info()
students_df.describe()
```

```python
students_df["average_score"] = (
    students_df["python_score"] + students_df["ml_score"]
) / 2

students_df
```

```python
high_performers = students_df[
    students_df["average_score"] >= 90
]

high_performers
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(students_df["student"], students_df["average_score"])
plt.title("Average Scores by Student")
plt.xlabel("Student")
plt.ylabel("Average Score")
plt.ylim(0, 100)
plt.show()
```

```python
students_df["study_efficiency"] = (
    students_df["average_score"] / students_df["hours_studied"]
)

students_df.sort_values(
    by="study_efficiency",
    ascending=False
)
```

## Key Takeaways

- Python collections organize structured information.
- Functions make code reusable and testable.
- NumPy performs efficient numerical operations.
- Pandas is used to inspect, transform, and analyze tabular data.
- Matplotlib converts data into visual information.
